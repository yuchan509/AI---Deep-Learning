```python
# 데이터 읽기.
df = pd.read_csv('data/ThoraricSurgery.csv', header=None)
x = df.drop(17, axis=1)
y = df[17]
```

```python
# 딥러닝 구조를 결정.( 모델을 설정하고, 실행하는 부분. )
model = Sequential()
model.add( Dense(30, input_dim=17, activation='relu') )
model.add( Dense(1, activation='sigmoid'))
```

```python
# 딥러닝 수행.
# 무작위성에 따른 성능 평가 오류 방지를 위해 seed값 설정.
tf.random.set_seed(1)
model.compile( loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x, y, epochs=100, batch_size=10)
```

```python
# 주의 사항.
# 예측 ( 결과가 1일 확률 )
# model.predict_classes(x) 문제가 있어 이 방법은 더 이상 사용하지 않게 됨.
pred = model.predict(x)   
pred
```

```python
# 위와 같은 방법을 권고.
result =( pred > 0.5 ).astype(int)
result
```

```python
# 모델 정확도 확인.
score = accuracy_score(y, result)
print(f'Accuracy : {round(score,3)}%')
```
