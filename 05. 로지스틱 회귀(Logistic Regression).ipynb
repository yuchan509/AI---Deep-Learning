```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터.
x_data = [2, 4, 6, 8, 10, 12, 14]

# 범주형으로 이루어져 있음.
y_data = [0, 0, 0, 1, 1, 1, 1]

print(x_data)
print(y_data)
```

```python
# 그래프.
plt.scatter(x_data, y_data, c='m', alpha=.5)
plt.grid()
plt.show()
```

```python
# a와 b의 값.
a = 0
b = 0

# 학습률.
lr = 0.05

# 반복 횟수.
epochs = 20001
```

```python
# 시그모이드 함수.
def sigmoid(x) :
    return 1 / (1 + np.e ** (-x))
```

```python
# ndarray 생성.
x_data2 = np.array(x_data)
y_data2 = np.array(y_data)

print(x_data2)
print(y_data2)
```

```python
for i in range(epochs) :
    # 예측 결과.
    y_pred = sigmoid(a + b * x_data2)
    # print(y_pred)
    
    # 오차 계산.
    error = y_pred - y_data2
    # print(error)
    
    # a와 b에 대한 편미분을 수행.
    b_diff = (1 / len(x_data2)) * sum(x_data2 * error)
    a_diff = (1 / len(x_data2)) * sum(error)
    # print(b_diff)
    # print(a_diff)
    
    # a와 b의 값을 계산.
    b = b - lr * b_diff
    a = a - lr * a_diff
    # print(b)
    # print(a)
    
    # 2000번 마다 결과 출력.
    if i % 2000 == 0 :
        print(f'epoch {i}, 절편 : {a}, 기울기 : {b}')
        plt.scatter(x_data2, y_data2, c='tomato', alpha=0.5)
        # 그래프로 나타낼 x값의 범위.
        x_range = (np.arange(0, 15, 0.1))
        plt.plot(x_range, np.array([sigmoid(a + b * x) for x in x_range]), c='m', alpha=0.5)
        plt.grid()
        plt.show()
```
